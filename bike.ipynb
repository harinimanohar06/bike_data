```python
df['Day']=df['Day'].fillna(df['Day']).mean()
```

```python
df['Order_Quantity']=df['Order_Quantity'].fillna(df['Order_Quantity']).median()
```

```python
df
```

```python
df["Age_Group"] = df["Age_Group"].fillna("Unknown")
df['Product_Description']=df['Age_Group'].fillna("Unknown")
```

```python
df.isna().sum()
```

```python
import numpy as np

ages = df["Customer_Age"].to_numpy()
mean_age = np.mean(ages)

print("Mean Customer Age:", mean_age)
```

```python
order_qty = df["Order_Quantity"].to_numpy()
mean_qty = np.mean(order_qty)
print("Mean Order Quantity:", mean_qty)
```

visualizations


```python
import matplotlib.pyplot as plt

plt.hist(df["Customer_Age"], bins=10, color="skyblue", edgecolor="black")
plt.title("Distribution of Customer Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns

sns.barplot(x="Age_Group", y="Revenue", data=df, estimator=sum, ci=None, palette="viridis")
plt.title("Total Revenue by Age Group")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby("Date")["Revenue"].sum().plot(kind="line", marker="o", figsize=(10,5))
plt.title("Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
age_counts = df["Age_Group"].value_counts()
product_counts = df["Product_Description"].value_counts()
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].pie(age_counts, labels=age_counts.index, autopct="%1.1f%%", startangle=90)
axes[0].set_title("Age Group")
axes[1].pie(product_counts, labels=product_counts.index, autopct="%1.1f%%", startangle=90)
axes[1].set_title("Product Description")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

search_counts = df["Age_Group"].value_counts()

search_counts.plot(kind="bar", color="lightgreen", edgecolor="black")
plt.title("Search Term Frequency")
plt.xlabel("Search Term")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("C:/Users/HARINI/Downloads/archive/uncleaned bike sales data.xlsx")
print("Available columns:", df.columns.tolist())

# Ask user for input
search_col = input("Enter the column to search (e.g., Country, Product_Description): ")
search_item = input("Enter the search term (e.g., United Kingdom, Mountain-200 Black): ")
if search_col in df.columns:
    filtered_df = df[df[search_col].str.contains(search_item, case=False, na=False)]
    print(f"Found {len(filtered_df)} records for '{search_item}' in column '{search_col}'")
    if not filtered_df.empty:
        revenue_by_product = filtered_df.groupby("Product_Description")["Revenue"].sum()
        plt.figure(figsize=(10,6))
        revenue_by_product.plot(kind="bar", color="skyblue", edgecolor="black")
        plt.title(f"Revenue by Product for {search_item}")
        plt.xlabel("Product Description")
        plt.ylabel("Total Revenue")
        plt.xticks(rotation=90)
        plt.show()
    else:
        print("No matching records found.")
else:
    print(f"Column '{search_col}' not found in dataset.")
```
